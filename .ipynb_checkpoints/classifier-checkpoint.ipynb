```python
from sklearn.model_selection import train_test_split
np.random.seed(42)
X = data.drop('target',axis=1)
Y = data.target

x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.2)
```

### Choosing the right model 

```python
np.random.seed(42)
models = {'RandomForestClassifier':RandomForestClassifier(),
          'LogisticRegression':LogisticRegression(max_iter=1000),
          'KNeighborsClassifier':KNeighborsClassifier()}

def eval_model(y_true,y_preds):
    acc = accuracy_score(y_true,y_preds)
    rec = recall_score(y_true,y_preds)
    pre = precision_score(y_true,y_preds)
    f1 = f1_score(y_true,y_preds)
    
    scores = {'accuracy':acc,
              'precision':pre,
              'recall':rec,
              'f1':f1}
    return scores
    
def best_baseline_model(models):
    for name,model in models.items():
        model.fit(x_train,y_train)
        y_preds = model.predict(x_test)
        score = eval_model(y_test,y_preds)
        print(name)
        for name,score in score.items():
            print(name,score)
        print("")
            
best_baseline_model(models)
```

### Hyperparameter tuning 

```python
# Tuning KNeighborsClassifier
np.random.seed(42)
neighbors = range(1,20)
modelk = KNeighborsClassifier()
scores = []
for i in neighbors:
    modelk.set_params(n_neighbors=i)
    modelk.fit(x_train,y_train)
    scores.append(modelk.score(x_test,y_test))
print(" Max score is :",max(scores)*100,'% for neighbors =',scores.index(max(scores))+1)
plt.plot(neighbors,scores);
plt.axvline(scores.index(max(scores))+1,color='salmon',label=' MAX score');
plt.legend();
```

```python
# Tuning Logistic Regression using GridSearchCV
np.random.seed(42)
logreg_grid = {'C': np.logspace(-4,4,20),
               'solver':['liblinear']}
model_log = LogisticRegression(n_jobs=1)
tuned_model_log = GridSearchCV(estimator=model_log,
                           param_grid=logreg_grid,
                           cv=5,
                           verbose=2)
tuned_model_log.fit(x_train,y_train)
```

```python
y_pred = tuned_model_log.predict(x_test)
eval_model(y_test,y_pred)
```

```python
tuned_model_log.best_params_
```

```python
# Tuning RandomForestClassifier using GridSearchCV
np.random.seed(42)
ran_grid = {'n_estimators': [500,1000],
       'max_depth': [30],
       'max_features': ["auto"],
       'min_samples_leaf': [1,2],
       'min_samples_split': [2,4]}
model_ran = RandomForestClassifier(n_jobs=1)
tuned_model_ran = GridSearchCV(estimator=model_ran,
                               param_grid=ran_grid,
                               cv=5,
                               verbose=2)
tuned_model_ran.fit(x_train,y_train)
```

```python
y_pred = tuned_model_ran.predict(x_test)
eval_model(y_test,y_pred)
```

### As our tuned LogisticRegression model is doing good in comparison to others 
* We will stick to it for the evaluation part

## Evaluating model

```python
# ROC curve
plt.style.use('seaborn-whitegrid')
RocCurveDisplay.from_estimator(tuned_model_log,x_test,y_test,color='green');
```

```python
# Evaluationg our model using confusion matrix
mat = confusion_matrix(y_test,tuned_model_log.predict(x_test))
def conf_mat(mat):
    fig,ax = plt.subplots(figsize=(6,5))
    ax = sns.heatmap(mat,annot=True,cmap='plasma')
    ax.set(xlabel='Predicted values',ylabel='True values')
y_pred = tuned_model_log.predict(x_test)
conf_mat(mat)
```

```python
# Classification report
print(classification_report(y_test,tuned_model_log.predict(x_test)))
```

```python
# Cross validated precision,recall,accuracy and f1 scores
def cross_val_report(clf,X,Y,n):
    acc = cross_val_score(clf,X,Y,scoring='accuracy',cv=n)
    pre = cross_val_score(clf,X,Y,scoring='precision',cv=n)
    rec = cross_val_score(clf,X,Y,scoring='recall',cv=n)
    f1 = cross_val_score(clf,X,Y,scoring='f1',cv=n)
    report = {'Accuracy':acc*100,'Precision':pre,'Recall':rec,'F1 score':f1}
    return report
report = cross_val_report(tuned_model_log,X,Y,5)    
```

```python
for name,scores in report.items():
    print(name,' Scores :',scores,' Mean :',scores.mean())
```
